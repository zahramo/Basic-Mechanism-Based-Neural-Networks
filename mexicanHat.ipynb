## Mexican Hat Implementation

```python
def do_mexican_hat_algorithm(s, R1, R2, C1, C2, t_max, x_max=2):
  if R2 == np.inf:
    R2 = len(s) - 1
  ## Step 0 
  w = [[0 for i in range(len(s))]for j in range(len(s))]
  results = {}
  for i in range(len(s)):
    for k in range(len(s)):
      if i+k < len(s):
        if k <= R1:
          w[i][i+k] = C1
        elif k > R1 and k <= R2:
          w[i][i+k] = C2
      if i-k > -1:
        if k <= R1:
          w[i][i-k] = C1
        elif k > R1 and k <= R2:
          w[i][i-k] = C2
  ## Step 1 
  x = np.array(s)
  t = 0
  x_old = x.copy()
  results[t] = x_old
  print('t = {}'.format(t))
  print('x = ')
  print(*x)
  t = 1
  ## Step 2 
  while(t < t_max):
    ## Step 3
    for i in range(len(x)):
      z = np.dot(x_old, w[i])
      ## Step 4
      if z<0: x[i] = 0
      elif 0<=z<x_max: x[i] = z
      elif z>x_max: x[i] = x_max
    ## Step 5
    x_old = x.copy()
    results[t] = x_old
    ## Step 6
    t += 1
    print('t = {}'.format(t))
    print('x = ')
    print(*x)
    max = s[do_max_net_algorithm(x_old, 0.13)]
    print('Max Number = {}'.format(max))
  return results, max
```

```python
def do_max_net_algorithm(input, e):
  ## Step 0
  num_count = len(input)
  weights = [[-e for i in range(num_count)] for j in range(num_count)]
  for j in range(num_count):
    for i in range(num_count):
      if i == j:
        weights[j][i] = 1
  A = np.array(input)
  print(*A)
  while(True):
    A_new = []
    for a in A:
      x = a - e*(A.sum()-a)
      if x >= 0: a_new = x
      else: a_new = 0
      A_new.append(a_new)
    A = np.array(A_new)
    print(*A)
    if  np.count_nonzero(A)<2:
      break
  return np.argmax(A, axis=None)
```

```python
def plot_results(results):
  plt.figure(figsize=(10,8))
  x_axis = range(len(results[0]))
  legends = []
  for t,result in results.items():
    plt.plot(x_axis, result)
    legends.append('t = {}'.format(t))
  x_label = 'Number Index'
  plt.title('Results for Mexican Hat')
  plt.xlabel(x_label)
  plt.legend(legends)
  plt.show()
```

```python
results, max = do_mexican_hat_algorithm([0.32, 0.33, 0.28, 0.47, 0.66, 0.8, 0.4, 0.33, 0.1, 0.26], 0, np.inf, 1.5, -0.2, 5)
plot_results(results)
```
